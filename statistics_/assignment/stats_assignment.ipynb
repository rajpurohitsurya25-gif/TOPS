3. Variance, Standard Deviation and Standard Error

```python
variance = np.var(df ["Flight Distance"])
print(f"variance of flight distance :- {variance}")
```

5. Find Outliers 


```python
mean = np.mean(df["Flight Distance"])
sd = np.std(df["Flight Distance"], ddof =1)

```

```python
from scipy import stats

stats.zscore(df["Flight Distance"])
```

```python
distances = df["Flight Distance"]
Q1 = np.percentile(distances,25)
Q3 = np.percentile(distances,75)
IQR = Q3 - Q1

lowerbond = Q1 - 1.5*IQR
upperbond = Q3 + 1.5*IQR

outliers = distances[(distances < lowerbond) | (distances > upperbond)]
print("Outliers:", outliers)
outliers.head()
```
